## Visualizing Earnings Based On College Majors


We'll be working with a dataset on the job outcomes of students who graduated from college between 2010 and 2012. The original data on job outcomes was released by American Community Survey(https://www.census.gov/programs-surveys/acs/), which conducts surveys and aggregates the data. FiveThirtyEight cleaned the dataset and released it on their Github repo(https://github.com/fivethirtyeight/data/tree/master/college-majors).

Each row in the dataset represents a different major in college and contains information on gender diversity, employment rates, median salaries, and more. Here are some of the columns in the dataset:

    Rank - Rank by median earnings (the dataset is ordered by this column).
    Major_code - Major code.
    Major - Major description.
    Major_category - Category of major.
    Total - Total number of people with major.
    Sample_size - Sample size (unweighted) of full-time.
    Men - Male graduates.
    Women - Female graduates.
    ShareWomen - Women as share of total.
    Employed - Number employed.
    Median - Median salary of full-time, year-round workers.
    Low_wage_jobs - Number in low-wage service jobs.
    Full_time - Number employed 35 hours or more.
    Part_time - Number employed less than 35 hours.
    
Using visualizations, we can start to explore questions from the dataset like:

- Do students in more popular majors make more money?
    - Using scatter plots
- How many majors are predominantly male? Predominantly female?
    - Using histograms
- Which category of majors have the most students?
    - Using bar plots

We'll explore how to do these and more while primarily working in pandas. Before we start creating data visualizations, let's import the libraries we need and remove rows containing null values.





```python
import pandas as pd
import matplotlib.pyplot as plt
% matplotlib inline
recent_grads = pd.read_csv('recent-grads.csv')
print(recent_grads.iloc[0])
print(recent_grads.head())
print(recent_grads.tail())
recent_grads = recent_grads.dropna()
```

- Generating scatter plots to explore the following relations:
    - Sample_size and Median
    - Sample_size and Unemployment_rate
    - Full_time and Median
    - ShareWomen and Unemployment_rate
    - Men and Median
    - Women and Median
- Using![alt text](https://) the plots to explore the following questions:
    - Do students in more popular majors make more money?
    - Do students that majored in subjects that were majority female make more money?
    - Is there any link between the number of full-time employees and median salary?

```python
fig = plt.figure(figsize=(12,12))
a1=fig.add_subplot(3,2,1)
a2=fig.add_subplot(3,2,2)
a3=fig.add_subplot(3,2,3)
a4=fig.add_subplot(3,2,4)
a5=fig.add_subplot(3,2,5)
a6=fig.add_subplot(3,2,6)
a1.scatter(recent_grads['Sample_size'], recent_grads['Median'])
a1.set_xlabel("sample sizes per major")
a1.set_ylabel("median salaries per major")
a1.set_title("Sample size vs. Median salary")
a2.scatter(recent_grads['Sample_size'], recent_grads['Unemployment_rate'])
a2.set_xlabel("sample sizes per major")
a2.set_ylabel("unemployment rates per major")
a2.set_title("Sample size vs. Unemployment rate ")
a3.scatter(recent_grads['Full_time'],recent_grads['Median'] )
a3.set_xlabel("full-timers per major")
a3.set_ylabel("median salaries per major")
a3.set_title("Full-timer count vs. Median salary")
a4.scatter(recent_grads['ShareWomen'], recent_grads['Unemployment_rate'])
a4.set_xlabel("women percentage")
a4.set_ylabel("unemployment rates per major")
a4.set_title("Women's share vs. Unemployment rate ")
a5.scatter(recent_grads['Men'],recent_grads['Median'])
a5.set_xlabel("Men population")
a5.set_ylabel("median salaries per major")
a5.set_title("Men population vs. Median salary")
a6.scatter(recent_grads['Women'],recent_grads['Median'])
a6.set_xlabel("Women population")
a6.set_ylabel("median salaries per major")
a6.set_title("Women population vs. Median salary")
plt.tight_layout()
```

- Generating histograms to explore the distributions of the following columns:
    - `Sample_size`
    - `Median`
    - `Employed`
    - `Full_time`
    - `ShareWomen`
    - `Unemployment_rate`
    - `Men`
    - `Women`
- We encourage you to experiment with different bin sizes and ranges when generating these histograms.
- Use the plots to explore the following questions:
    - What percent of majors are predominantly male? Predominantly female?
    - What's the most common median salary range?

```python
cols = ["Sample_size", "Median", "Employed", "Full_time", "ShareWomen", "Unemployment_rate", "Men", "Women"]

fig = plt.figure(figsize=(5,12))
for r in range(1,5):
    ax = fig.add_subplot(2,2,r)
    ax = recent_grads[cols[r-1]].plot(kind='hist', rot=40)
```

```python
cols = ["Sample_size", "Median", "Employed", "Full_time", "ShareWomen", "Unemployment_rate", "Men", "Women"]

fig = plt.figure(figsize=(5,12))
for r in range(4,8):
    ax = fig.add_subplot(2,2,r-3)
    ax = recent_grads[cols[r]].plot(kind='hist', rot=40)
```

In the last 2 steps, we created individual scatter plots to visualize potential relationships between columns and histograms to visualize the distributions of individual columns. A scatter matrix plot combines both scatter plots and histograms into one grid of plots and allows us to explore potential relationships and distributions simultaneously. 
We are exploring the scatter matrix plots on our variables to study the underlying distributions and relationships
- Creating a 2 by 2 scatter matrix plot using the Sample_size and Median columns.
- Creating a 3 by 3 scatter matrix plot using the Sample_size, Median, and Unemployment_rate columns.
-Creating 3 by 3 scatter matrix plots to understand the Median salary and unemployment distribution accross the genders
- Explore the questions from the last few steps using these scatter matrix plots. 

```python
from pandas.plotting import scatter_matrix
scatter_matrix(recent_grads[['Sample_size', 'Median']], figsize=(6,6))
```

```python
scatter_matrix(recent_grads[['Sample_size', 'Median', 'Unemployment_rate']], figsize=(10,10))
```

```python
scatter_matrix(recent_grads[['Men', 'Median', 'Unemployment_rate']], figsize=(10,10))
```

```python
scatter_matrix(recent_grads[['Women', 'Median', 'Unemployment_rate']], figsize=(10,10))
```

From the above scatter matrices, it is evident that the distribution of populations of men and women with respect to salaries and unemloyment rates are similar.
